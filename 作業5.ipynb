```python
#y=ax^2+bx+c  x=1~100 or 500 求鋒值,x軸交點
#1.a=-1,b=4,c=5
#2.a=3,b=2,c=-6
```

```python
a=-1
b=4
c=5
for x in range(1,101):
  y=a*(x**2)+(b*x)+c
  print(x,y)
print("與X的交點(5,0)和(-1,0)")
d=-b//2*a
e=(4*a*c-b**2)//4*a
print("峰值為",d,",",e)
```
