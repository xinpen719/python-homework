```python
#y=ax^2+bx+c  x=1~100 or 500 求鋒值,x軸交點
#1.a=-3,b=4,c=7
#2.a=5,b=-4,c=9
```

```python
a=-3
b=4
c=7
for x in range(1,101):
  y=a*(x**2)+(b*x)+c
  print(x,y)
print("與X的交點(10,0)和(-10,0)")
d=-b//2*a
e=(4*a*c-b**2)//4*a
print("峰值為",d,",",e)


import matplotlib.pyplot as plt
import numpy as np
 
n = 100
x = np.linspace(-5, 5, n) 
y = y=a*(x**2)+(b*x)+c
  
plt.plot(x, y)
plt.grid(True)
plt.xlabel('X') 
plt.ylabel('Y')
plt.title('y=a*(x**2)+(b*x)+c')
plt.show()  
```
